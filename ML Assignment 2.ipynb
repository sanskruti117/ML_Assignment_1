```python
# To learn Basic Linear Regression
import numpy as np
from sklearn.linear_model import LinearRegression
x=np.array([5,15,25,35,45,55]).reshape((-1,1))
y=np.array([5,20,14,32,22,38])
print(x)
print(y)
```

```python
model=LinearRegression().fit(x,y)
r_sq=model.score(x,y)
print("Coffecient of Determination:",r_sq)
print("Intercept:",model.intercept_)
print("Slope:",model.coef_)
```

```python
y_pred=model.predict(x)
print("Predicted response:",y_pred)
```

```python
df.describe()
```

```python
# Data distribution
plt.scatter(df['YearsExperience'], df['Salary'], color='purple')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.box(False)
plt.show()
```

```python
X = df.iloc[:, :1] 
y = df.iloc[:, 1:] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
linearModel= LinearRegression()
linearModel.fit(X_train, y_train)
```

```python
y_pred_test = linearModel.predict(X_test)    
y_pred_train = linearModel.predict(X_train)
```

```python
plt.scatter(X_train, y_train, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='pink')
plt.show()
```

```python
print(f'Coefficient: {linearModel.coef_}')
print(f'Intercept: {linearModel.intercept_}')
```

```python
r_sq1 = linearModel.score(X, y)
print(f"coefficient of determination: {r_sq1}")
```
